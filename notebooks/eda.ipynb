# Data exploration

```python
from src.ingest import fetch_s3_data

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Using our predefinedfunction to fetch the data
df = fetch_s3_data("staywise-data-bucket/airbnb/raw_data", "AB_NYC_2019.csv")

df.head()
```

### Target variable ('price')

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], bins=100, kde=True)
plt.title('Distribution of Airbnb Prices in NYC')
plt.show()
```

```python
df['price'].describe()
```

**Observations:** 
75% of listings are under $175

```python
plt.figure(figsize=(10, 2))
sns.boxplot(x=df['price'])
plt.title('Identifying Price Outliers')
plt.show()
```

### Price by neighbourhood

```python
# Filtering to listings under $500 for better visibility
df_filtered = df[df['price'] <= 500]

plt.figure(figsize=(12, 6))
sns.violinplot(x='neighbourhood_group', y='price', data=df_filtered)
plt.title('Price Distribution by Neighbourhood Group (Listings <= $500)')
plt.show()
```

**Observations:**
* Manhattan has the highest median price and the widest upper distribution, with a noticeable bulge around $150–$300, confirming its premium market position
* Queens, Staten Island, and Bronx all display similar and relatively low median prices, clustered tightly near $50–$100

### Price by room type

```python
# Using the filtered data (price <= 500) for clarity
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df_filtered)
plt.title('Price Distribution by Room Type (Listings <= $500)')
plt.show()
```

**Observations:**
* Entire home/apt has the highest median price at approximately $150–$160 and shared room has the lowest median at approximately $45–$50
* Private room has a narrow IQR (~$50–$100), meaning most listings are consistently budget-friendly with little spread
* Private room also has by far the most outliers, with a dense cloud of dots extending from ~$160 all the way to $500+, suggesting many atypical high-priced private rooms
